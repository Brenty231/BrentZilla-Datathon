Rank players by WAR(Wins Above Replacement):

```python
df.sort_values('WAR:', ascending=False)
```

With the exception of one clear outlier(a pitcher), there is a positive correlation between Wins Above Replacement and voting rank
We will investigate if a stronger correlation exists between WAR and only position players later on 

```python
plt.scatter(df['WAR:'], df.index)
plt.xlabel('WAR') 
plt.ylabel('Voting Rank')
plt.yticks([0.0, 20.0],['First Place', 'Last Place'])
plt.title('WAR/Ranking Correlation:')
plt.gca().invert_yaxis()
```

There also seems to be a correlation between OPS(on-base percentage slugging percentage)

```python
plt.scatter(df['OPS:'], df.index)
plt.xlabel('OPS') 
plt.ylabel('Voting Rank')
plt.yticks([0.0, 20.0],['First Place', 'Last Place'])
plt.title('OPS/Ranking Correlation:')
plt.gca().invert_yaxis()
```

```python
cols=list(df.columns)
```

Rank players first by OPS, then by WAR:

```python
df.sort_values(['OPS:', 'WAR:'], ascending=False)
```

Judging from above, it seems as if OPS(for position players) and WAR(for both position players and pitchers) are two heavily weighted statistics used by the BBWAA in voting

What statistics hold the most weight for pitchers?:

Sort by Wins and Innings Pitched(the higher, the better):

```python
df.sort_values(['IP:','W:'], ascending=False)
```

WHIP is the average walks and hits allowed per inning pitched. ERA is the average earned runs allowed per 9 innings pitched(the lower, the better)

```python
df.sort_values(['WHIP:','ERA:'])
```

```python
plt.scatter(df['WHIP:'], df['Vote Pts:'])
plt.xlabel('WHIP')
plt.ylabel('Voting Points')
plt.title('Pitcher WHIP/Voting Rank Correlation')
plt.gca().invert_xaxis()
```

```python
plt.scatter(df['ERA:'], df['Vote Pts:'])
plt.xlabel('ERA')
plt.ylabel('Voting Points')
plt.title('Pitcher ERA/Voting Rank Correlation')
plt.gca().invert_xaxis()
```

Judging from above, WHIP and ERA seem to have more wegiht than Wins and Innings Pitched
