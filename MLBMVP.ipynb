While the WAR statistic has a clear positive correlation for position players, the same does not hold for pithcers, why is this?

WAR values reflect a higher contribution to a team's success. Since, WAR integrates more areas of the game (hitting, fielding, baserunning, and pitching) and position players play in substantially more games than pitchers, their WAR values are typically weighted heavier. Position players play in substantially more games, so it is more difficult for them to maintain a high WAR value over the course of the season.

Now let's look at the correlations for pitchers

First of all, there are four statistical categories for pitchers not included in this list, Strikeouts per 9 innings pitched, Strikeouts to Walks Ratio, Innings Pitched per game started, and Fielding Independent pitching(FIP). Let's add those categories to this dataframe

Fielding Independent Pitching(FIP) is a complex stat used to measure a pitchers performance independent of defense. The coefficients for it's equation, (13, 3, 2), are determined by the league average of batting average on balls in play(BABIP). Some pitchers have better defense behind them, so in short, FIP does a good job of measuring a pitcher's performance with average defensive players around him. The FIP constant Constant is determined by an average ERA independent of defense(only includes walks, strikeouts, and home runs). Calculating the precise constant would require a stat list of all MLB pitchers. The historical average of FIP is between 3.1 and 3.2, so we will use 3.15.

In simpler terms, FIP is ERA(Average Earned Runs allowed per 9 innings pitched) with a team's defensive strength irrelevant

```python
df_1['SO/9:']=9*(df_1['SO:']/df_1['IP:'])
df_1['SO/BB:']=df_1['SO:']/df_1['BB:.1']
df_1['IP/GS:']=df_1['IP:']/df_1['GS:']
df_1['FIP:']= (((df_1['HR:.1']*13)+(df_1['BB:.1']*3)-(df_1['SO:']*2))/df_1['IP:'])+3.15
df_1
```
